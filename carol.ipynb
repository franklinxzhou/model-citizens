```python
train_set['driver_age'] = train_set_eda['year'] - train_set_eda['year_of_born']
pd.crosstab(train_set['subrogation'], train_set['driver_age'])
```

```python
len(train_set[(train_set['driver_age'] >= 14) & (train_set['driver_age'] <= 120)])
```

```python
train_set['driving_experience'] = (train_set['driver_age'] - train_set['age_of_DL']).clip(lower=0)
pd.crosstab(train_set['subrogation'], train_set['driving_experience'])
```

```python
train_set['month'] = pd.to_datetime(train_set['claim_date']).dt.month
train_set['is_weekend'] = train_set['claim_day_of_week'].isin(['Saturday', 'Sunday']).astype(int)
pd.crosstab(train_set['subrogation'], train_set['is_weekend'])
```

```python
season_map = {
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall',
    12: 'Winter', 1: 'Winter', 2: 'Winter'
}

train_set['season'] = train_set['month'].map(season_map)
pd.crosstab(train_set['subrogation'], train_set['season'])
```

```python
train_set['log_claim_est_payout'] = np.log(train_set['claim_est_payout'])
train_set['log_vehicle_mileage'] = np.log(train_set['vehicle_mileage'])
train_set['log_vehicle_price'] = np.log(train_set['vehicle_price'])
```

```python
train_set['income_to_vehicle_price'] = train_set['annual_income'] / (1 + train_set['vehicle_price'])
train_set['safety_x_prior_claims'] = train_set['safety_rating'] / (1 + train_set['past_num_of_claims'])
train_set['experience_x_safety'] = train_set['driving_experience'] * train_set['safety_rating']
train_set['liab_x_witness'] = train_set['liab_prct'] * train_set['witness_present_ind'].map({'Y':1, 'N':0})
```

```python
test_set['year'] = pd.to_datetime(test_set['claim_date']).dt.year
test_set['month'] = pd.to_datetime(test_set['claim_date']).dt.month
test_set['driver_age'] = test_set['year'] - test_set['year_of_born']
test_set['driving_experience'] = (test_set['driver_age'] - test_set['age_of_DL']).clip(lower=0)
test_set['is_weekend'] = test_set['claim_day_of_week'].isin(['Saturday', 'Sunday']).astype(int)
test_set['season'] = test_set['month'].map(season_map)

test_set['log_claim_est_payout'] = np.log(test_set['claim_est_payout'])
test_set['log_vehicle_mileage'] = np.log(test_set['vehicle_mileage'])
test_set['log_vehicle_price'] = np.log(test_set['vehicle_price'])

test_set['income_to_vehicle_price'] = test_set['annual_income'] / (1 + test_set['vehicle_price'])
test_set['safety_x_prior_claims'] = test_set['safety_rating'] / (1 + test_set['past_num_of_claims'])
test_set['experience_x_safety'] = test_set['driving_experience'] * test_set['safety_rating']
test_set['liab_x_witness'] = test_set['liab_prct'] * test_set['witness_present_ind'].map({'Y':1, 'N':0})
```

# Logistic Regression

## First Try

```python
X_train = train_set.iloc[:, 1:]
y_train = train_set.iloc[:, 0]
```

```python
X_train['gender'] = np.where(X_train['gender'] == 'F', 1, 0)
X_train['living_status'] = np.where(X_train['living_status'] == 'Own', 1, 0)
```

```python
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns
num_cols = X_train.select_dtypes(exclude=['object', 'category']).columns

# Preprocessor
preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
        ('num', StandardScaler(), num_cols)]
)

# Model pipeline
model = Pipeline([
    ('preprocess', preprocess),
    ('logreg', LogisticRegression(
        solver='lbfgs',
        max_iter=5000,
        random_state=SEED,
        class_weight='balanced'  # good for imbalanced data
    ))
])

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(test_set)
```

## Second Try

```python
train_set_filtered = train_set.copy()
#train_set_filtered = train_set_filtered[(train_set_filtered['driver_age'] >= 14) & (train_set_filtered['driver_age'] <= 120)]
#train_set_filtered[['driver_age', 'driving_experience']].describe(include='all')
```

```python
X_train = train_set_filtered.iloc[:, 1:]
X_train = X_train.drop(columns = ['claim_number', 'vehicle_made_year', 'claim_date'])

y_train = train_set_filtered.iloc[:, 0]
```

```python
X_train_train, X_train_test, y_train_train, y_train_test = train_test_split(X_train, y_train, test_size=0.3, random_state=SEED)
```

```python
cat_cols = X_train_train.select_dtypes(include=['object', 'category']).columns
num_cols = X_train_train.select_dtypes(exclude=['object', 'category']).columns

# Preprocessor
preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
        ('num', StandardScaler(), num_cols)]
)

# Model pipeline
model = Pipeline([
    ('preprocess', preprocess),
    ('logreg', LogisticRegression(
        solver='lbfgs',
        max_iter=5000,
        random_state=SEED,
        class_weight='balanced'  # good for imbalanced data
    ))
])

model.fit(X_train_train, y_train_train)
```

```python
y_pred = model.predict(X_train_test)
```

```python
accuracy = accuracy_score(y_train_test, y_pred)
print(f"Model Accuracy: {accuracy}")
```

### Feature Selection

```python
preprocess = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ]), cat_cols),

        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), num_cols)
    ]
)

lasso_selector = SelectFromModel(
    LogisticRegression(
        penalty='l1',
        solver='saga',
        max_iter=5000,
        class_weight='balanced',
        random_state=SEED,
        C=0.1  # strength of regularization (lower = stronger feature elimination)
    )
)

final_model = LogisticRegression(
    solver='lbfgs',
    max_iter=5000,
    class_weight='balanced',
    random_state=SEED
)

pipeline_lasso = Pipeline([
    ('preprocess', preprocess),
    ('lasso_select', lasso_selector),
    ('model', final_model)
])

pipeline_lasso.fit(X_train_train, y_train_train)
```

```python
feature_mask = pipeline_lasso.named_steps['lasso_select'].get_support()
feature_names = pipeline_lasso.named_steps['preprocess'].get_feature_names_out()
selected_features = np.array(feature_names)[feature_mask]

print(selected_features)
```

```python
y_pred = pipeline_lasso.predict(X_train_test)
```

```python
accuracy = accuracy_score(y_train_test, y_pred)
print(f"Model Accuracy: {accuracy}")
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_train_test, y_pred)
print("F1 Score:", f1)
```
